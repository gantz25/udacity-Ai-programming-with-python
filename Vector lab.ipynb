```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# creating matrix [1,1]
v=np.array([1,1])
# creating axis of plot referenced as ax
ax=plt.axes()

#making red dot at the origin 
ax.plot(0,0,'or')

# plotting a v vector at 00
ax.arrow(0,0,*v,color='b',linewidth=2,head_width=.2,head_length=.25)
plt.xlim(-2,2)
# setting major ticks for x axis

major_ticks=np.arange(-3,2)
ax.set_xticks(major_ticks)

# setting limit for y axis
plt.ylim(-1,2)

major_yticks=np.arange(-1,3)

ax.set_yticks(major_yticks)

plt.grid(b=True,which='major')

plt.show()
```

```python
# scaling a vector
a=3
av=a*v
ax=plt.axes()
ax.plot(0,0,'or')
ax.arrow(0,0,*v,linewidth=2.5,head_width=.3,head_length=.3)
ax.arrow(0,0,*av,color='c',linestyle='dotted',head_width=.3,head_length=.3)
plt.xlim(-2,4)
major_xticks=np.arange(-2,4)
plt.ylim(-1,5)
major_yticks=np.arange(-1,5)
ax.set_yticks(major_yticks)
ax.set_xticks(major_xticks)
plt.grid(b=True,which='major')
plt.show()
```

```python
# adding two vectors
#plotting result of v[1 1] and w=[-2 2]
w=np.array([-2,2])
ax=plt.axes()
ax.plot(0,0,'or')
ax.arrow(0,0,*v,color='b',linewidth=2.5,head_width=.3,head_length=.3)
ax.arrow(0,0,*w,color='c',linewidth=2.5,head_width=.3,head_length=.3)
plt.xlim(-4,4)
plt.ylim(-1,6)
major_yticks=np.arange(-1,6)
major_xticks=np.arange(-4,4)
ax.set_xticks(major_xticks)
ax.set_yticks(major_yticks)
plt.grid(b=True,which='major')
plt.show()


```

```python
#vector addition
ax=plt.axes()
ax.plot(0,0,'or')
wv=w+v
ax.arrow(0,0,*v,color='b',linewidth=2.5,head_width=.3,head_length=.3)
ax.arrow(v[0],v[1],*w,linestyle='dotted',linewidth=2.5,head_width=.3,head_length=.3)
plt.xlim(-4.3)
plt.ylim(-1,5)
major_xticks=np.arange(-4,3)
major_yticks=np.arange(-1,5)
ax.set_xticks(major_xticks)
ax.set_yticks(major_yticks)
plt.grid(b=True,which='major')
plt.show();
```

```python
# plotting the result of the addition
vw=w+v
ax=plt.axes()
ax.plot(0,0,'or')
ax.arrow(0,0,*vw,color='k',linewidth=2.5,head_width=.3,head_length=.3)
ax.arrow(0,0,*v,color='b',linewidth=2.5,head_width=.3,head_length=.3)
ax.arrow(v[0],v[1],*w,linestyle='dotted',linewidth=2.5,head_width=.3,head_length=.3)
major_xticks=np.arange(-3,2)
major_yticks=np.arange(-1,4)
plt.xlim(-3,2)
plt.ylim(-1,4)
ax.set_xticks(major_xticks)
ax.set_yticks(major_yticks)
plt.grid(b=True,which='major')
plt.show();
```
